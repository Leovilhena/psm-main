Suggest and run a computer application simulating the dynamic system drawing the Lissajous curve.

```python
from datetime import datetime
from numpy import sin,pi,linspace
from matplotlib.pylab import plot, show, subplot, savefig, subplots
from matplotlib.animation import FuncAnimation

a = [1,3,5,3] # plotting the curves for
b = [1,5,7,4] # different values of a/b
delta = pi/2  # Î´
t = linspace(-pi,pi,300)

def animate(i):
    line.set_data(x[:i], y[:i])
    return line,
```

```python
for i, (a, b) in enumerate(zip(a, b)):
    # Lissajous curve graph equation
    x = sin(a*t + delta)  # x = Asin(at+Î´)
    y = sin(b*t)  # y = Bsin(bt)
    # plot x and y
    fig, ax = subplots()
    line, = ax.plot(x, y)

    show()
    # save GIF animation
    anim = FuncAnimation(fig, animate, frames=len(x),interval=25, blit=True)
    anim.save(f'lissajous_curve-{i}.gif', writer='pillow')
```
